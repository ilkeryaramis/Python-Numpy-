# NumPy Operations

```python
import numpy as np
```

```python
arr= np.arange(0,11)
```

```python
arr+arr     #element wise sumation
```

```python
arr*arr     # element wise product
```

```python
arr *5    # scaler operation
```

```python
arr /arr   # Warning on division by zero, but not an error!
```

```python
1./arr     # Also warning, but not an error instead infinity
```

```python
arr** 3
```

## Universal Array Functions

```python
np.sqrt(arr)     # takes sqrt of all elements in array
```

```python
np.exp(arr)
```

```python
np.log(arr)
```

```python
np.sin(arr)
```
